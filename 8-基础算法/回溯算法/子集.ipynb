给你一个整数数组 nums ，数组中的元素互不相同.返回该数组所有可能的子集（幂集）。

输入：nums = [1,2,3]

输出：[[],[1],[2],[1,2],[3],[1,3],[2,3],[1,2,3]]

````python
class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        res = []   # 结果集：存放所有子集
        path = []  # 当前路径：正在构建的子集

        def backtracking(index: int):
            # 1. 每次进入函数，当前path就是一个子集，加入结果
            res.append(path[:])  # 使用path[:]创建副本，避免后续修改影响结果;如果不创建副本：res.append(path) 会添加引用
            
            # 2. 递归终止条件：所有元素都考虑过了
            if index >= len(nums):
                return
                
            # 3. 从当前下标开始，尝试每个可能的元素
            for i in range(index, len(nums)): # 从index开始遍历，保证了不会重复考虑之前已经处理过的元素，避免了产生重复子集
                path.append(nums[i])    # 选择：将nums[i]加入当前子集
                backtracking(i + 1)     # 递归：考虑下一个元素
                path.pop()               # 撤销选择：回溯到上一步

        backtracking(0)  # 从第一个元素开始
        return res
````

以[1, 2, 3]为例
```text
                backtracking(0)  path=[]
                    |
        +----------+----------+----------+
        |          |          |          |
    i=0选1      i=1选2      i=2选3    i=3不选
    path=[1]   path=[2]   path=[3]   res=[[]]
        |          |          |
    backtrack(1) backtrack(2) backtrack(3)
        |          |
    i=1选2      i=2选3
    path=[1,2]  path=[2,3]
        |          |
    backtrack(2) backtrack(3)
        |
    i=2选3
    path=[1,2,3]
        |
    backtrack(3)
```
