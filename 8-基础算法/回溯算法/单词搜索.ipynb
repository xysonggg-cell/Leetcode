给定一个 m x n 二维字符网格 board 和一个字符串单词 word 。如果 word 存在于网格中，返回 true ；否则，返回 false 。

单词必须按照字母顺序，通过相邻的单元格内的字母构成，其中“相邻”单元格是那些水平相邻或垂直相邻的单元格。同一个单元格内的字母不允许被重复使用。

>输入：board = [['A','B','C','E'],['S','F','C','S'],['A','D','E','E']], word = "ABCCED"
输出：true

```python
class Solution:
    def exist(self, board: List[List[str]], word: str) -> bool:
        # 四个方向：右、左、下、上
        directs = [(0, 1), (0, -1), (1, 0), (-1, 0)]
        rows = len(board)
        if rows == 0:
            return False
        cols = len(board[0])

        # 记录已访问的单元格
        visited = [[False for _ in range(cols)] for _ in range(rows)]

        def backtrack(i, j, index):
            # 终止条件：已匹配到最后一个字符
            if index == len(word) - 1:
                # 检查当前单元格是否等于最后一个字符
                return board[i][j] == word[index]

            # 如果当前字符匹配
            if board[i][j] == word[index]:
                # 标记当前位置已访问
                visited[i][j] = True

                # 向四个方向探索
                for direct in directs:
                    new_i = i + direct[0]
                    new_j = j + direct[1]
                    # 检查边界和是否已访问
                    if 0 <= new_i < rows and 0 <= new_j < cols and not visited[new_i][new_j]:
                        # 递归探索下一个字符
                        if backtrack(new_i, new_j, index + 1):
                            return True

                # 回溯：撤销标记
                visited[i][j] = False

            return False

        # 从每个单元格开始尝试
        for i in range(rows):
            for j in range(cols):
                if backtrack(i, j, 0):
                    return True
        return False
```

$backtrack(i, j, index)$ 表示从$ board[i][j]$ 出发，能否搜索到单词字母 $word[index]$，以及 $index$ 位置之后的后缀子串。如果能搜索到，则返回 $True$，否则返回 $False$

对每一个位置 (i,j) 都调用函数 check(i,j,0) 进行检查：只要有一处返回 true，就说明网格中能够找到相应的单词，否则说明不能找到
