按照国际象棋的规则，皇后可以攻击与之处在同一行或同一列或同一斜线上的棋子。
n 皇后问题研究的是如何将 n 个皇后放置在 n×n 的棋盘上，并且使皇后彼此之间不能相互攻击。

给你一个整数 n ，返回所有不同的 n 皇后问题的解决方案。

每一种解法包含一个不同的 n 皇后问题的棋子放置方案，该方案中 'Q' 和 '.' 分别代表了皇后和空位

输入：n = 4

输出：[[".Q..","...Q","Q...","..Q."],["..Q.","Q...","...Q",".Q.."]]


````python
class Solution:
    res = []  # 类变量，存储所有有效解
    
    def solveNQueens(self, n: int) -> List[List[str]]:
        self.res.clear()  # 清空之前的结果（避免累积）
        
        # 初始化棋盘，全部填充 '.'
        chessboard = []
        for i in range(n):
            temp = []
            for j in range(n):
                temp.append('.')  # '.' 表示空位
            chessboard.append(temp)  # 将每一行加入棋盘
        
        # 从第0行开始回溯
        self.backtracking(0, chessboard, n)
        return self.res
    
    def isValid(self, n, row, col, chessboard):
        # 1. 检查同一列是否有皇后（只需要检查当前行以上的行）
        for i in range(row):
            if chessboard[i][col] == 'Q':
                return False
        
        # 2. 检查左上对角线
        i, j = row - 1, col - 1
        while i >= 0 and j >= 0:
            if chessboard[i][j] == "Q":
                return False
            i -= 1
            j -= 1
        
        # 3. 检查右上对角线
        i, j = row - 1, col + 1
        while i >= 0 and j < n:
            if chessboard[i][j] == "Q":
                return False
            i -= 1
            j += 1
        
        # 所有检查都通过，这个位置可以放皇后
        return True
    
    def backtracking(self, row, chessboard, n):
        # 终止条件：所有行都放置了皇后（找到一组解）
        if row >= n:
            # 将棋盘格式化为题目要求的输出格式
            temp_res = []
            for temp in chessboard:
                temp_res.append(''.join(temp))  # 将每行从列表转为字符串
            self.res.append(temp_res)  # 加入结果集
            return
        
        # 尝试在当前行的每一列放置皇后
        for col in range(n):
            if self.isValid(n, row, col, chessboard):
                # 选择：在(row, col)放置皇后
                chessboard[row][col] = 'Q'
                
                # 递归：处理下一行
                self.backtracking(row + 1, chessboard, n)
                
                # 撤销选择：回溯，尝试下一列
                chessboard[row][col] = '.'
````

```python
# 撤销选择：回溯，尝试下一列
chessboard[row][col] = '.'
```

个人对回溯的理解：

如果不用回溯的话，就需要双循环，复杂度$O(n^2)$。用回溯的话，`self.backtracking(row + 1, chessboard, n)`就相当于循环，同时减少时间复杂度
