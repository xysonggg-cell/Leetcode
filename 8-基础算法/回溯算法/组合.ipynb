给定两个整数 n 和 k，返回范围 [1, n] 中所有可能的 k 个数的组合。

你可以按 任何顺序 返回答案。

>输入：n = 4, k = 2  
>输出：
>[
  [2,4],
  [3,4],
  [2,3],
  [1,2],
  [1,3],
  [1,4],
]

````python
class Solution:
    def combine(self, n: int, k: int) -> List[List[int]]:
        res = []
        
        def backtrack(start, combination):
            # 终止条件：组合长度达到 k
            if len(combination) == k:
                res.append(combination[:])  # 添加副本；如果不使用副本，后续修改 combination 会影响已添加的结果
                return # 结束当前递归分支!!!
            
            # 从 start 开始，避免重复
            for i in range(start, n + 1):
                combination.append(i)                # 做选择
                backtrack(i + 1, combination)        # 递归（从 i+1 开始，避免重复）
                combination.pop()                    # 撤销选择
        
        backtrack(1, [])
        return res
````

```backtrack(1, [])
├── i=1: [1]
│   ├── backtrack(2, [1])
│   │   ├── i=2: [1,2] → len==2 → 记录 → return ✅ (不再继续添加)
│   │   ├── i=3: [1,3] → 记录 → return ✅
│   │   └── i=4: [1,4] → 记录 → return ✅
│   └── 循环结束，自然返回
├── i=2: [2]
│   └── backtrack(3, [2])
│       ├── i=3: [2,3] → 记录 → return ✅
│       └── i=4: [2,4] → 记录 → return ✅
└── i=3: [3]
    └── backtrack(4, [3])
        └── i=4: [3,4] → 记录 → return ✅
