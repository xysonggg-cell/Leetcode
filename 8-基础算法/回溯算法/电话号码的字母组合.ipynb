给定一个仅包含数字 2-9 的字符串，返回所有它能表示的字母组合。答案可以按 任意顺序 返回。

给出数字到字母的映射如下（与电话按键相同）。注意 1 不对应任何字母。

>输入：digits = "23"  
>输出：["ad","ae","af","bd","be","bf","cd","ce","cf"]

```python
class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
        # 边界情况：空字符串返回空列表
        if not digits:
            return []

        # 数字到字母的映射字典（电话键盘布局）
        phone_dict = {
            "2": "abc", "3": "def", "4": "ghi",
            "5": "jkl", "6": "mno", "7": "pqrs",
            "8": "tuv", "9": "wxyz"
        }

        res = []  # 存储所有结果
        
        def backtrack(combination, index):
            # 终止条件：已经处理完所有数字
            if index == len(digits):
                res.append(combination)  # 将当前组合加入结果
            else:
                # 获取当前数字对应的字母串
                digit = digits[index]
                # 遍历每个可能的字母
                for i in phone_dict[digit]:
                    # 递归：选择当前字母，处理下一个数字
                    # 显示回溯：<<<在递归调用前添加当前字母，递归调用后移除当前字母>>>
                    combination += i
                    backtrack(combination, index + 1)
                    combination = combination[:-1]  # 回溯：移除当前字母

        
        backtrack('', 0)  # 从空字符串和第0个数字开始
        return res
```

体会回溯的流程：和dfs很像，先深入搜索，到达终点之后回退
