数字 n 代表生成括号的对数，请你设计一个函数，用于能够生成所有可能的并且 有效的 括号组合。

>输入：n = 3
输出：["((()))","(()())","(())()","()(())","()()()"]

```python
class Solution:
    def generateParenthesis(self, n: int) -> List[str]:
        parentheses = []            # 存放所有有效的括号组合
        parenthesis = []            # 存放当前正在构建的括号组合（字符列表）

        def backtrack(symbol, index):
            # symbol: 当前未匹配的左括号数量（即还可以放多少个右括号）
            # index: 当前已经放置的括号总数

            # 终止条件：已经放置了 2n 个括号
            if n * 2 == index:
                # 如果所有左括号都已匹配（symbol == 0），说明是一个有效组合
                if symbol == 0:
                    parentheses.append("".join(parenthesis))
            else:
                # 情况1：还可以放左括号（已放左括号数 < n）
                if symbol < n:
                    parenthesis.append('(')
                    backtrack(symbol + 1, index + 1)  # symbol+1：未匹配左括号数+1
                    parenthesis.pop()  # 回溯

                # 情况2：还可以放右括号（有未匹配的左括号）
                if symbol > 0:
                    parenthesis.append(')')
                    backtrack(symbol - 1, index + 1)  # symbol-1：未匹配左括号数-1
                    parenthesis.pop()  # 回溯

        backtrack(0, 0)  # 从 0 个未匹配左括号、0 个总括号开始
        return parentheses
```
