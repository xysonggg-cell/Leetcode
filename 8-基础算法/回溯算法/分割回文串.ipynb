给你一个字符串 s，请你将 s 分割成一些 子串，使每个子串都是 回文串 。返回 s 所有可能的分割方案

>输入：s = "aab"
输出：[["a","a","b"],["aa","b"]]

```python
class Solution:
    res = []   # 存储所有分割方案
    path = []  # 存储当前分割方案

    def backtrack(self, s: str, start_index: int):
        # 终止条件：已经处理完整个字符串
        if start_index >= len(s):
            self.res.append(self.path[:])  # 添加副本，避免后续修改
            return

        # 从 start_index 开始，尝试所有可能的结束位置
        for i in range(start_index, len(s)):
            # 如果子串 s[start_index:i+1] 是回文
            if self.ispalindrome(s, start_index, i):
                # 选择：将回文子串加入当前路径
                self.path.append(s[start_index: i+1])

                # 递归：处理剩余部分
                self.backtrack(s, i + 1)

                # 撤销选择：回溯
                self.path.pop()

    def ispalindrome(self, s: str, start: int, end: int):
        # 双指针判断是否为回文
        i, j = start, end
        while i < j:
            if s[i] != s[j]:
                return False
            i += 1
            j -= 1
        return True

    def partition(self, s: str) -> List[List[str]]:
        # 清空上一轮的结果（因为 res 和 path 是类变量）
        self.res.clear()
        self.path.clear()
        self.backtrack(s, 0)
        return self.res
```

使用回溯算法，在字符串的每个位置尝试切割：
1. 从起始位置开始，尝试所有可能的结束位置
2. 如果当前子串是回文，则将其加入路径
3. 递归处理剩余部分
4. 当处理完整个字符串时，记录当前分割方案
