实现 pow(x, n) ，即计算 x 的整数 n 次幂函数（即，x^n )

````python
class Solution:
    def myPow(self, x: float, n: int) -> float:
        # 定义递归函数，只处理非负指数
        def quickMul(N):
            # 递归终止条件：任何数的0次方都是1
            if N == 0:
                return 1.0
            
            # 递归计算 x^(N//2)
            y = quickMul(N // 2)
            
            # 根据N的奇偶性组合结果
            if N % 2 == 0:
                # 偶数情况：x^N = (x^(N/2))^2
                return y * y
            else:
                # 奇数情况：x^N = x × (x^((N-1)/2))^2
                return x * y * y
        
        # 处理正负指数
        # 如果n>=0，直接计算x^n
        # 如果n<0，计算x^(-n)再取倒数
        return quickMul(n) if n >= 0 else 1.0 / quickMul(-n)
````

数学原理：

$x^n$ 可以分解为：
- 如果 n 是偶数：$x^n = (x^{(n/2)})*(x^{(n/2)})$
- 如果 n 是奇数：$x^n = x × (x^{((n-1)/2)})^2$

```text
quickMul(5)
    ├─ N=5 奇数 → 返回 2 * y * y
    │   y = quickMul(2)
    │       ├─ N=2 偶数 → 返回 y * y
    │       │   y = quickMul(1)
    │       │       ├─ N=1 奇数 → 返回 2 * y * y
    │       │       │   y = quickMul(0)
    │       │       │       └─ N=0 返回 1.0
    │       │       │   y = 1.0
    │       │       │   返回 2 * 1.0 * 1.0 = 2
    │       │       y = 2
    │       │       返回 2 * 2 = 4
    │       y = 4
    │       返回 2 * 4 * 4 = 32
结果：32
```

合理采用幂的相乘，时间复杂度$O(log n)$
