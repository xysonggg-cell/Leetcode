在一条环路上有 n 个加油站，其中第 i 个加油站有汽油 gas[i] 升。

你有一辆油箱容量无限的的汽车，从第 i 个加油站开往第 i+1 个加油站需要消耗汽油 cost[i] 升。你从其中的一个加油站出发，开始时油箱为空。

给定两个整数数组 gas 和 cost ，如果你可以按顺序绕环路行驶一周，则返回出发时加油站的编号，否则返回 -1 。如果存在解，则 保证 它是 唯一 的。

>输入: gas = [1,2,3,4,5], cost = [3,4,5,1,2]  
输出: 3

### 方法一：暴力（超时）

```python
class Solution:
    def canCompleteCircuit(self, gas: List[int], cost: List[int]) -> int:
        n = len(gas)
        # 拼接数组处理循环（也可以使用取模）
        gas2 = gas + gas
        cost2 = cost + cost
        
        for i in range(n):
            cur = 0
            valid = True
            for j in range(i, i + n):
                cur += gas2[j] - cost2[j]
                if cur < 0:
                    valid = False
                    break
            if valid:
                return i
        return -1
```

### 方法二：贪心算法

```python
class Solution:
    def canCompleteCircuit(self, gas: List[int], cost: List[int]) -> int:
        n = len(gas)
        total_tank = 0
        curr_tank = 0
        start = 0
        
        for i in range(n):
            total_tank += gas[i] - cost[i]
            curr_tank += gas[i] - cost[i]
            
            # 如果当前油量不足，无法到达下一站
            if curr_tank < 0:
                start = i + 1  # 从下一站重新开始；不是 start + 1，因为如果从 start 出发，到 j 时油量变负，那么从 start+1 到 j 的任何点出发，也一定会在到达 j 之前油量变负
                curr_tank = 0  # 重置当前油量
        
        # 如果总油量大于等于总消耗，一定有解
        return start if total_tank >= 0 else -1
```

总油量 >= 总消耗  ←→  存在解

如果总油量够，一定存在某个起点
