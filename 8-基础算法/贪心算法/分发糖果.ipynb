n 个孩子站成一排。给你一个整数数组 ratings 表示每个孩子的评分。

你需要按照以下要求，给这些孩子分发糖果：

+ 每个孩子至少分配到 1 个糖果。
+ 相邻两个孩子中，评分更高的那个会获得更多的糖果。
+ 请你给每个孩子分发糖果，计算并返回需要准备的 最少糖果数目 

```python
class Solution:
    def candy(self, ratings: List[int]) -> int:
        num = [1 for i in range(len(ratings))]

        for i in range(1, len(ratings)):
            if ratings[i] > ratings[i - 1]:
                num[i] = num[i - 1] + 1
        for j in range(len(ratings) - 2, -1, -1): # 第二遍循环是反着来的
            if ratings[j] > ratings[j + 1]:
                num[j] = max(num[j], num[j + 1] + 1)

        total = sum(num)
        
        return total
```

两次遍历的原因：

+ 第一遍（左→右）：保证每个孩子比左边评分高的孩子糖果多

+ 第二遍（右→左）：保证每个孩子比右边评分高的孩子糖果多，同时不破坏第一遍的结果

倒着循环的原因：

+ 让右边邻居的糖果数已经确定，才能正确更新左边

+ 保证信息的单向传递（从右向左递推）
