给定一个区间的集合 intervals ，其中 intervals[i] = [starti, endi] 。返回需要移除区间的最小数量，使剩余区间互不重叠 。

注意只在一点上接触的区间是不重叠的。例如 [1, 2] 和 [2, 3] 是不重叠的。

输入: intervals = [[1,2],[2,3],[3,4],[1,3]]

输出: 1


```python
class Solution:
    def eraseOverlapIntervals(self, intervals: List[List[int]]) -> int:
        if not intervals:
            return 0
        
        intervals.sort(key = lambda x: x[1])
        end_pos = intervals[0][1]
        count = 1
        for i in range(1, len(intervals)):
            if end_pos <= intervals[i][0]:
                count += 1
                end_pos = intervals[i][1]
                
        return len(intervals) - count
```

移除最少的区间，使得剩下的区间互不重叠。等价于：找到最多数量的互不重叠区间

贪心策略：按区间结束时间排序，每次都选择结束最早的区间，选择结束最早的区间可以保证不重叠区间数量最多
+ 局部最优选择：在可选区间中，优先选择结束时间最早的
+ 全局最优结果：最终得到最多数量的互不重叠区间
