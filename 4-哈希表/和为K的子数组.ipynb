给你一个整数数组 nums 和一个整数 k ，请你统计并返回 该数组中和为 k 的子数组的个数 。

子数组是数组中元素的连续非空序列。

>输入：nums = [1,1,1], k = 2
>输出：2

>输入：nums = [1,2,3], k = 3
输出：2

```python
class Solution:
    def subarraySum(self, nums: List[int], k: int) -> int:
        # 哈希表：key 是前缀和，value 是该前缀和出现的次数
        pre_dic = {0: 1}  # 初始化：前缀和为 0 出现 1 次（空子数组）
        pre_sum = 0  # 当前前缀和
        count = 0    # 满足条件的子数组个数

        for num in nums:
            pre_sum += num  # 更新当前前缀和

            # 检查是否存在前缀和等于 pre_sum - k
            # 如果存在，说明从这些前缀和的下一个位置到当前位置的子数组和为 k
            if pre_sum - k in pre_dic:
                count += pre_dic[pre_sum - k]

            # 记录当前前缀和出现的次数
            if pre_sum in pre_dic:
                pre_dic[pre_sum] += 1
            else:
                pre_dic[pre_sum] = 1

        return count
```
