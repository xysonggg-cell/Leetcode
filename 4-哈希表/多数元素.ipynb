给定一个大小为 n 的数组 nums ，返回其中的多数元素。多数元素是指在数组中出现次数大于⌊ n/2 ⌋ 的元素。

输入：nums = [2,2,1,1,1,2,2]

输出：2

```python
class Solution:
    def majorityElement(self, nums: List[int]) -> int:
        hashtable = dict()
        i = 0
        while i <= len(nums) - 1:
            if nums[i] not in hashtable:
                hashtable[nums[i]] = 1
            else:
                hashtable[nums[i]] += 1
            i += 1

        max = 0
        maxIndex = -1
        for num in hashtable:
            if hashtable[num] > max:
                max = hashtable[num]
                maxIndex = num
        return maxIndex
```
