给定一个已排序的链表的头 head ，删除所有重复的元素，使每个元素只出现一次。返回已排序的链表。

输入：head = [1,1,2,3,3]

输出：[1,2,3]

```python
class Solution:
    def deleteDuplicates(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if head is None:
            return None

        p_slow = head
        p_fast = head

        while p_fast is not None:
            if p_fast.val != p_slow.val:
                p_slow.next = p_fast
                p_slow = p_slow.next
                p_fast = p_fast.next
            else:
                p_fast = p_fast.next
                
            p_slow.next = None # 这行很关键，把slow后面的都断开。
                               # 不能放到if语句中，这样的话[1,1]测试案例无法通过

        return head
```

这道题和“删除有序数组的重复元素”思路一样，额外考察链表的应用
