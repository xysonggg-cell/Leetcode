给你一个按照非递减顺序排列的整数数组 nums，和一个目标值 target。请你找出给定目标值在数组中的开始位置和结束位置。

如果数组中不存在目标值 target，返回 [-1, -1]。

你必须设计并实现时间复杂度为 O(log n) 的算法解决此问题。

>输入：nums = [5,7,7,8,8,10], target = 8
>输出：[3,4]

### 方法一

```python
class Solution:
    def searchRange(self, nums: List[int], target: int) -> List[int]:
        left = 0
        right = len(nums) - 1
        first = -1
        last = -1
        # 找第一个等于target的位置
        while left <= right:
            mid = left + (right - left) // 2
            if nums[mid] == target:
                first = mid
                right = mid - 1  # 重点！！！
            elif nums[mid] > target:
                right = mid - 1
            else:
                left = mid + 1

        # 最后一个等于target的位置
        left = 0
        right = len(nums) - 1
        while left <= right:
            mid = left + (right - left) // 2
            if nums[mid] == target:
                last = mid
                left = mid + 1  # 重点！！！
            elif nums[mid] > target:
                right = mid - 1
            else:
                left = mid + 1

        return [first, last]
```

### 方法二

```python
class Solution:
    def searchRange(self, nums: List[int], target: int) -> List[int]:
        ans = [-1, -1]
        n = len(nums)
        if n == 0:
            return ans

        # ========== 第一次二分：找左边界 ==========
        left = 0
        right = n - 1
        while left < right:
            mid = left + (right - left) // 2  # 偏左的中点
            if nums[mid] < target:
                left = mid + 1               # 目标在右边
            else:
                right = mid                  # 目标在左边（包括 mid）

        # 检查是否找到目标
        if nums[left] != target:
            return ans
        ans[0] = left  # 记录左边界

        # ========== 第二次二分：找右边界 ==========
        left = 0
        right = n - 1
        while left < right:
            mid = left + (right - left + 1) // 2  # 偏右的中点
            if nums[mid] > target:
                right = mid - 1               # 目标在左边
            else:
                left = mid                   # 目标在右边（包括 mid）

        ans[1] = left  # 记录右边界
        return ans
```
