```python
# 自己的方法，比较麻烦
class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        sort_str = [sorted(i) for i in strs]
        mp = collections.defaultdict(list)
        for i in range(len(strs)):
            mp[''.join(sort_str[i])].append(strs[i])
        
        return list(mp.values())
```

```python
# 标准答案的方法
class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        mp = collections.defaultdict(list)

        for str in strs:
            mp[''.join(sorted(str))].append(str)

        return list(mp.values())
```

对字符串排序可以用`sorted(str)`

`collections.defaultdict(list)`: 得到的是值为列表的字典
