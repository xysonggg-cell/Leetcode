给定一个包含红色、白色和蓝色、共 n 个元素的数组 nums ，原地 对它们进行排序，使得相同颜色的元素相邻，并按照红色、白色、蓝色顺序排列。

我们使用整数 0、 1 和 2 分别表示红色、白色和蓝色。

必须在不使用库内置的 sort 函数的情况下解决这个问题。

>输入：nums = [2,0,2,1,1,0]
输出：[0,0,1,1,2,2]

### 方法一：单指针

```python
class Solution:
    def sortColors(self, nums: List[int]) -> None:
        """
        Do not return anything, modify nums in-place instead.
        """
        ptr = 0
        for i in range(len(nums)):
            if nums[i] == 0:
                nums[ptr], nums[i] = nums[i], nums[ptr]
                ptr += 1
        for i in range(len(nums)):
            if nums[i] == 1:
                nums[ptr], nums[i] = nums[i], nums[ptr]
                ptr += 1

        return nums
```

ptr：表示数组 nums 从位置 0 到位置 ptr−1 都属于「头部」

在第一次遍历中，从左向右遍历整个数组，如果找到了 0，那么就需要将 0 与「头部」位置的元素进行交换，并将「头部」向后扩充一个位置。
在第二次遍历中，从左向右遍历整个数组，如果找到了 1，那么就需要将 1 与「头部」位置的元素进行交换，并将「头部」向后扩充一个位置。


### 方法二：双指针

```python
class Solution:
    def sortColors(self, nums: List[int]) -> None:
        left = 0                    # 0 的右边界（下一个0应该放的位置）
        right = len(nums) - 1       # 2 的左边界（下一个2应该放的位置）
        index = 0                   # 当前遍历位置

        while index <= right:
            if nums[index] == 0:
                # 将 0 交换到 left 位置
                nums[index], nums[left] = nums[left], nums[index]
                left += 1
                index += 1
            elif nums[index] == 2:
                # 将 2 交换到 right 位置
                nums[index], nums[right] = nums[right], nums[index]
                right -= 1
                # 注意：index 不移动，因为交换过来的数还需要检查
            else:  # nums[index] == 1
                # 1 在正确的位置，继续前进
                index += 1
```
