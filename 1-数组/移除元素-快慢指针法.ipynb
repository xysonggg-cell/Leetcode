```python
class Solution:
    def removeElement(self, nums: List[int], val: int) -> int:
        slow = 0
        fast = 0
        while fast < len(nums):
            if nums[fast] != val:
                nums[slow] = nums[fast]
                fast += 1
                slow += 1
            else:
                fast += 1
        return slow
```

逻辑：
如果 fast 遇到值为 val 的元素，则直接跳过，

如果 fast 没遇到值为 val 的元素，就赋值给 slow 指针，并让 slow 前进一步
