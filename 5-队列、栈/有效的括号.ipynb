```python
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []

        for item in s:
            # 如果是左括号，把对应的右括号压入栈
            if item == '(':
                stack.append(')')
            elif item == '[':
                stack.append(']')
            elif item == '{':
                stack.append('}')

            # 如果是右括号，检查是否匹配
            elif not stack or item != stack[-1]:
                # 情况1: not stack → 栈为空，但遇到了右括号（如 "]"）
                # 情况2: stack[-1] != item → 栈顶括号与当前右括号不匹配
                return False 
            else:
                stack.pop()
        return True if not stack else False
```

这道题分三种情况考虑

1. 能匹配上，但是右括号比左括号多
2. 遇到的括号无法匹配
3. 能匹配上，但是左括号比右括号多
