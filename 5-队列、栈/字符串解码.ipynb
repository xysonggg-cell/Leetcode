给定一个经过编码的字符串，返回它解码后的字符串。

编码规则为: k[encoded_string]，表示其中方括号内部的 encoded_string 正好重复 k 次。注意 k 保证为正整数。

你可以认为输入字符串总是有效的；输入字符串中没有额外的空格，且输入的方括号总是符合格式要求的。

此外，你可以认为原始数据不包含数字，所有的数字只表示重复的次数 k ，例如不会出现像 3a 或 2[4] 的输入。

测试用例保证输出的长度不会超过 105。

>输入：s = "3[a]2[bc]"
>输出："aaabcbc"

>输入：s = "3[a2[c]]"
>输出："accaccacc"

```python
class Solution:
    def decodeString(self, s: str) -> str:
        stack1 = []  # 存储当前层之前的解码结果
        stack2 = []  # 存储当前层的重复次数
        num = 0      # 当前解析的数字
        res = ""     # 当前层的解码结果

        for ch in s:
            if ch.isdigit():
                # 处理数字（可能有多位，如 "12[a]"）
                num = num * 10 + int(ch)

            elif ch == '[':
                # 遇到 '['，将当前状态入栈，开始新的一层
                stack1.append(res)  # 保存之前的结果
                stack2.append(num)  # 保存重复次数
                res = ""            # 重置当前结果
                num = 0             # 重置数字

            elif ch == ']':
                # 遇到 ']'，结束当前层，出栈并重复
                cur_res = stack1.pop()   # 之前的结果
                cur_num = stack2.pop()   # 重复次数
                res = cur_res + res * cur_num  # 拼接结果

            else:
                # 普通字符，直接加入当前结果
                res += ch

        return res
```
