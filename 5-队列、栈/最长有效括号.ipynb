### 方法一：栈

给你一个只包含 '(' 和 ')' 的字符串，找出最长有效（格式正确且连续）括号 子串 的长度。

左右括号匹配，即每个左括号都有对应的右括号将其闭合的字符串是格式正确的，比如 "(()())"。

>输入：s = "(()"
输出：2

```python
class Solution:
    def longestValidParentheses(self, s: str) -> int:
        stack = [-1]
        ans = 0
        for i in range(len(s)):
            if s[i] == "(":
                stack.append(i)
            if s[i] == ")":
                stack.pop()
                if not stack:
                    stack.append(i)
                else:
                    ans = max(ans, i - stack[-1])
        return ans
```

初始先在栈中加入-1作为哨兵节点，表示 最长有效括号子串的开始元素的前一个元素下标为-1

如果stack pop之后为空，说明之前弹出的栈顶元素实际上是「最长有效括号子串的开始元素的前一个元素下标」
