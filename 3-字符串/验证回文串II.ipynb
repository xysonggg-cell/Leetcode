给定一个非空字符串 s，请判断如果最多从字符串中删除一个字符能否得到一个回文字符串

```python
# 方法一 自己的方法，但是超时了
class Solution:
    def validPalindrome(self, s: str) -> bool:
        if self.check(s):
            return True
        
        for i in range(len(s)):
            new_s = s[:i] + s[i + 1:]
            if self.check(new_s):
                return True
        return False

    def check(self, s):
        left = 0
        right = len(s) - 1
        while left < right:
            if s[left] != s[right]:
                return False
            left += 1
            right -= 1
        return True
```

```python
# 方法二，借助递归
class Solution:
    def validPalindrome(self, s: str) -> bool:
        def check(left, right):
            while left < right:
                if s[left] != s[right]:
                    return False
                left += 1
                right -= 1
            return True

        left = 0
        right = len(s) - 1
        while left < right:
            if s[left] == s[right]:
                left += 1
                right -= 1
            else:
                return check(left + 1, right) or check(left, right - 1)
        return True

    
```

依次比较左右指针对应的字符是否相等。
+ 如果相等，继续比较剩下的字符。
+ 如果不相等，则分两种情况，只要有一种情况是回文字符串即可：
  + 删除左边的 left 指针指向的元素，判断 s[left+1, right] 是否回文。
  + 删除右边的 right 指针指向的元素，判断 s[left, right-1] 是否回文。
