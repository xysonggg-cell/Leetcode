给定一个字符串 s ，验证 s 是否是回文串，只考虑字母和数字字符，可以忽略字母的大小写

```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        if s == '':
            return True

        res = [i.lower() for i in s if i.isalnum()]
        
        left = 0
        right = len(res) - 1
        while left < right:
            if res[left] != res[right]:
                return False
            left += 1
            right -= 1
        
        return True
        
```
