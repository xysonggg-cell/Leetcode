```python
class Solution:
    def preorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        res = []

        def traverse(root):  # 这里相当于写一个遍历的函数，然后再调用函数
            if not root:
                return
            traverse(root.left)
            res.append(root.val)
            traverse(root.right)

        traverse(root)
        return res
```

中序遍历：【对于一个节点，如果它的左子节点已经被遍历完了，但是还没有遍历右子节点】，按照这样的顺序进行的遍历
