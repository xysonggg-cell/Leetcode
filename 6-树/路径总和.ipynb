给你二叉树的根节点 root 和一个表示目标和的整数 targetSum 。判断该树中是否存在 根节点到叶子节点 的路径，这条路径上所有节点值相加等于目标和 targetSum 。如果存在，返回 true ；否则，返回 false 。

叶子节点 是指没有子节点的节点。

### 方法一

```python
class Solution:
    def hasPathSum(self, root: Optional[TreeNode], targetSum: int) -> bool:
        return self.sum(root, targetSum, 0)
    
    def sum(self, root, targetSum, curSum):
        if not root:
            return False
        
        curSum += root.val
        
        if not root.left and not root.right:
            return curSum == targetSum
        
        return self.sum(root.left, targetSum, curSum) or self.sum(root.right, targetSum, curSum)
```

### 方法二

```python
class Solution:
    def hasPathSum(self, root: Optional[TreeNode], targetSum: int) -> bool:
        if not root:
            return False
        if not root.left and not root.right:
            return root.val == targetSum # Check if the leaf node's value equals the remaining target sum
        
        return self.hasPathSum(root.left, targetSum - root.val) or self.hasPathSum(root.right, targetSum - root.val)
```
