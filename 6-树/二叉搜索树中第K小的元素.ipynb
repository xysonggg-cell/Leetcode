给定一个二叉搜索树的根节点 root ，和一个整数 k ，请你设计一个算法查找其中第 k 小的元素（k 从 1 开始计数）。

```python
# 自己想的方法
class Solution:
    def kthSmallest(self, root: Optional[TreeNode], k: int) -> int:
        self.arr = []
        def traverse(root):
            if not root:
                return

            traverse(root.left)
            self.arr.append(root.val)
            traverse(root.right)
        
        traverse(root)
        return self.arr[k - 1]
```

直接用中序遍历得到递增的数组
