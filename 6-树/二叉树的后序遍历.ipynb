```python
class Solution:
    def preorderTraversal(self, root: Optional[TreeNode]) -> List[int]:
        res = []

        def traverse(root):  # 这里相当于写一个遍历的函数，然后再调用函数
            if not root:
                return
            traverse(root.left)
            traverse(root.right)
            res.append(root.val)

        traverse(root)
        return res
```

一个节点的所有左右子节点全部都遍历完了，就把这个节点放入列表res中
