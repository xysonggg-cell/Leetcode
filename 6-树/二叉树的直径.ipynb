给你一棵二叉树的根节点，返回该树的 直径 。

二叉树的 直径 是指树中任意两个节点之间最长路径的 长度 。这条路径可能经过也可能不经过根节点 root 。

两节点之间路径的 长度 由它们之间边数表示。

````python
class Solution:
    def __init__(self):
        # 记录最大直径的长度（全局变量）
        self.maxDiameter = 0

    def diameterOfBinaryTree(self, root):
        # 调用maxDepth计算深度，过程中会更新maxDiameter
        self.maxDepth(root)
        # 返回计算好的最大直径
        return self.maxDiameter

    def maxDepth(self, root):
        # 基础情况：空节点深度为0
        if root is None:
            return 0
        
        # 递归计算左右子树的深度
        leftMax = self.maxDepth(root.left)
        rightMax = self.maxDepth(root.right)
        
        # 【关键】后序位置：此时已经知道左右子树的最大深度
        # 经过当前节点的直径 = 左子树深度 + 右子树深度
        myDiameter = leftMax + rightMax
        # 更新全局最大直径
        self.maxDiameter = max(self.maxDiameter, myDiameter)

        # 返回当前节点的深度（给父节点用）
        return 1 + max(leftMax, rightMax) # eg. 节点1的左最大深度=2节点的深度，所以要返回
````

计算二叉树的直径 -> 可以想到借助二叉树的最大深度

第二遍做的时候没有写`return 0`和`return 1 + max(leftMax, rightMax)`,如果不写这两句的话就没有办法返回数值。  
之前遍历二叉树的时候不写是因为不需要返回数值，但最大深度需要返回数值，所以需要一个基准值（0）和增加的值（每个节点的深度）

下面示例可以很形象地说明为什么要返回数值
```text
        1
       / \
      2   3
     / \
    4   5
```
```text
（因为是后序遍历，所以要从最后一个往前推）
maxDepth(4) → left=0, right=0 → maxDiameter=0 → 返回 1
maxDepth(5) → left=0, right=0 → maxDiameter=0 → 返回 1
maxDepth(2) → left=1, right=1 → maxDiameter=2 → 返回 2
maxDepth(3) → left=0, right=0 → maxDiameter=2 → 返回 1
maxDepth(1) → left=2, right=1 → maxDiameter=3 → 返回 3
```
