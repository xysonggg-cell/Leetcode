给你二叉树的根节点 root ，返回其节点值的层序遍历

eg.

输入：root = [3,9,20,null,null,15,7]

输出：[[3],[9,20],[15,7]]

```python
class Solution:
    def levelOrder(self, root: Optional[TreeNode]) -> List[List[int]]:
        res = []
        if not root:
            return [] # 如果root是空树的话返回空列表
        q = collections.deque() # 使用deque双向队列
        q.append(root)
        # deep = 1

        while q:
            length = len(q)
            temp = []
            for i in range(length):
                cur = q.popleft() # 双向队列的特有函数popleft
                temp.append(cur.val)


                if cur.left:
                    q.append(cur.left)
                if cur.right:
                    q.append(cur.right)
            # deep += 1
            res.append(temp)
        
        return res
```
