给你一个链表，删除链表的倒数第 n 个结点，并且返回链表的头结点。

>输入：head = [1,2,3,4,5], n = 2
输出：[1,2,3,5]

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
        size = 0
        p = head
        while p:
            size += 1
            p = p.next

        # k考虑size = 1的特殊情况
        if size == 1:
            return None

        q = head
        # 如果n = size，删除第一个节点
        if size - n - 1 < 0:
            return head.next

        for i in range(size - n - 1):
            q = q.next
        # 删除最后一个节点
        if not q.next.next:
            q.next = None
            return head
        q.next = q.next.next

        return head
```
