```python
class Solution:
    def mergeTwoLists(self, list1: Optional[ListNode], list2: Optional[ListNode]) -> Optional[ListNode]:
        
        dummy = ListNode(-1)  # 创建一个虚拟头节点，值为-1   [-1 | None]
        p1 = list1  # p1指针指向第一个链表
        '''
        p1 -----> [1 | *] ----> [3 | *] ----> [5 | None]
                   ↑
                list1指向这里
        '''
        p2 = list2
        p = dummy  # p指针指向新链表的当前节点


        while p1 is not None and p2 is not None:
            if p1.val <= p2.val:
                p.next = p1
                p1 = p1.next
            elif p1.val > p2.val:
                p.next = p2
                p2 = p2.next
            p = p.next


        while p1 is not None:
            p.next = p1
            p1 = p1.next
            p = p.next
        

        while p2 is not None:
            p.next = p2
            p2 = p2.next
            p = p.next

        return dummy.next
        # return dummy 会得到-1 -> 1 -> 2 -> 3 -> 4 -> 5 -> 6
        # dummy.next才会得到最后的结果
```
