给你一个链表的头节点 head 和一个特定值 x ，请你对链表进行分隔，使得所有小于 x 的节点都出现在大于或等于 x 的节点之前。

输入：head = [1,4,3,2,5,2], x = 3

输出：[1,2,2,4,3,5]

````python
# 方法一 
class Solution:
    def partition(self, head: Optional[ListNode], x: int) -> Optional[ListNode]:
        dummy1 = ListNode(-1)
        dummy2 = ListNode(-1)
        p1 = dummy1
        p2 = dummy2
        p = head

        while p is not None:
            if p.val < x:
                p1.next = p
                p1 = p1.next
            else:
                p2.next = p
                p2 = p2.next
            temp = p.next  # 这里是想把p和p前面的断开，等同于把p1和p1后面的断开
            p.next = None
            p = temp  
        p1.next = dummy2.next
        return dummy1.next
````

````python
# 方法二
class Solution:
    def partition(self, head: Optional[ListNode], x: int) -> Optional[ListNode]:
        dummy1 = ListNode(-1)
        dummy2 = ListNode(-1)
        p1 = dummy1
        p2 = dummy2
        p = head

        while p is not None:
            if p.val < x:
                p1.next = p
                p1 = p1.next
                p = p.next
                p1.next = None # 因为p1后面连接了p后面的全部，所以要把p1后面断开
            else:
                p2.next = p
                p2 = p2.next
                p = p.next
                p2.next = None
            
        p1.next = dummy2.next
        return dummy1.next
````
