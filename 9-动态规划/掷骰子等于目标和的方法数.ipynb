这里有 n 个一样的骰子，每个骰子上都有 k 个面，分别标号为 1 到 k 。

给定三个整数 n、k 和 target，请返回投掷骰子的所有可能得到的结果（共有 kn 种方式），使得骰子面朝上的数字总和等于 target。

由于答案可能很大，你需要对 109 + 7 取模。

```python
class Solution:
    def numRollsToTarget(self, n: int, k: int, target: int) -> int:
        # 初始化 DP 数组，长度为 target+1
        dp = [0 for _ in range(target + 1)]
        dp[0] = 1  # 0 个骰子，总和为 0 的方案数为 1
        MOD = 10 ** 9 + 7  # 结果取模，防止溢出

        # 枚举前 i 个骰子
        for i in range(1, n + 1):
            # 逆序枚举背包容量（避免重复使用同一个骰子）
            for w in range(target, -1, -1):
                dp[w] = 0  # 重新计算当前骰子数下的方案数
                # 枚举当前骰子可能掷出的点数
                for d in range(1, k + 1):
                    if w >= d:
                        dp[w] = (dp[w] + dp[w - d]) % MOD
        
        return dp[target] % MOD
```
