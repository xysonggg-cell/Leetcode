给你一个整数数组 nums ，请你找出一个具有最大和的连续子数组（子数组最少包含一个元素），返回其最大和。

子数组是数组中的一个连续部分。

输入：nums = [-2,1,-3,4,-1,2,1,-5,4]

输出：6

解释：连续子数组 [4,-1,2,1] 的和最大，为 6 。

```python
# 自己的方法，超时，时间复杂度O(n^3)
class Solution:
    def maxSubArray(self, nums: List[int]) -> int:
        dp = [0 for _ in range(len(nums))]
        if len(nums) == 1:
            return nums[0]

        for i in range(len(nums)):
            total = []
            for j in range(i + 1):
                total.append(sum(nums[j: i + 1]))
                dp[i] = max(total)

        return max(dp)
```

```python
# 标准方法
class Solution:
    def maxSubArray(self, nums: List[int]) -> int:
        size = len(nums)
        dp = [0 for _ in range(size)]
        
        dp[0] = nums[0]  # 以第一个元素结尾的最大子数组和就是它本身
        
        for i in range(1, size):
            if dp[i - 1] < 0:      # 如果前面的累加和是负数
                dp[i] = nums[i]     # 重新开始，从当前元素起算
            else:                    # 如果前面的累加和是非负数
                dp[i] = dp[i - 1] + nums[i]  # 接在后面的子数组上
        
        return max(dp)  # 所有以 i 结尾的最大子数组和中的最大值
```

设 $dp[i]$ 表示以第 i 个元素结尾的连续子数组的最大和

状态转移方程：
+ 如果 $dp[i−1]<0$，说明以$ i−1$ 结尾的子数组对当前元素有负贡献，此时不如从当前元素重新开始，即 $dp[i]=nums[i]$。
+ 如果 $dp[i−1]≥0$，则以 $i−1$ 结尾的子数组对当前元素有正贡献，可以将其累加，即$dp[i]=dp[i−1]+nums[i]$。
