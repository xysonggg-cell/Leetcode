给你一个整数 n ，返回 和为 n 的完全平方数的最少数量 。

```python
class Solution:
    def numSquares(self, n: int) -> int:
        weight = [i ** 2 for i in range(1, n) if i ** 2 <= n]
        dp = [n + 1 for _ in range(n + 1)]
        dp[0] = 0

        if n == 1:
            return 1

        for i in range(len(weight)):
            for w in range(weight[i], n + 1):
                dp[w] = min(dp[w], dp[w - weight[i]] + 1)

        return dp[n]
```

看作完全背包问题

将问题转换成：在完全平方数中选择最少数量的数，使它们的和等于n

定义状态：  
$dp[w]$：从完全平方数中挑选一些数，使其和恰好凑成 $w$ ，最少需要多少个完全平方数  
状态转移方程：  
$dp[w]=min(dp[w],dp[w−num]+1)$

在上面的代码中，weight是完全平方数数组（可以看成是不同的石头），n可以看作W（背包的最大容量）
