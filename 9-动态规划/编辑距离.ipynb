给你两个单词 word1 和 word2， 请返回将 word1 转换成 word2 所使用的最少操作数  。

你可以对一个单词进行如下三种操作：
+ 插入一个字符
+ 删除一个字符
+ 替换一个字符

>输入：word1 = "horse", word2 = "ros"  
>输出：3  
>解释：  
>horse -> rorse (将 'h' 替换为 'r')  
>rorse -> rose (删除 'r')  
>rose -> ros (删除 'e')  

```python
class Solution:
    def minDistance(self, word1: str, word2: str) -> int:
        size1 = len(word1)
        size2 = len(word2)
        
        # 创建 (size1+1) x (size2+1) 的 DP 表
        dp = [[0 for _ in range(size2 + 1)] for _ in range(size1 + 1)]

        # 初始化第一列：将 word1 的前 i 个字符转换为空字符串，需要 i 次删除
        for i in range(size1 + 1):
            dp[i][0] = i
        
        # 初始化第一行：将空字符串转换为 word2 的前 j 个字符，需要 j 次插入
        for j in range(size2 + 1):
            dp[0][j] = j
        
        # 填充 DP 表
        for i in range(1, size1 + 1):
            for j in range(1, size2 + 1):
                # 如果当前字符相等，不需要操作
                if word1[i - 1] == word2[j - 1]:
                    dp[i][j] = dp[i - 1][j - 1]
                else:
                    # 取三种操作的最小值，然后 +1
                    dp[i][j] = min(
                        dp[i - 1][j - 1],  # 替换：将 word1[i] 替换为 word2[j]
                        dp[i - 1][j],      # 删除：删除 word1[i]
                        dp[i][j - 1]       # 插入：在 word1 中插入 word2[j]
                    ) + 1
        
        # 返回将整个 word1 转换为整个 word2 的最少操作次数
        return dp[size1][size2]
```

定义状态：  
$dp[i][j]$ 表示将 $word1[:i]$ 转换成 $word2[:j]$ 的编辑距离

状态转移方程：  
如果当前字符相同（$word1[i−1]==word2[j−1]$），则无需操作：$dp[i][j]=dp[i−1][j−1]$  
如果当前字符不同（$word1[i−1]!=word2[j−1]$），则有三种操作可选，取最小值：  
+ 替换：$dp[i−1][j−1]+1$
+ 插入：$dp[i][j−1]+1$
+ 删除：$dp[i−1][j]+1$

替换：  
将 $word1[i-1]$ 替换成 $word2[j-1]$。  
第一步：将 $word1[:i-1]$ 变成 $word2[:j-1]$   -->   $dp[i−1][j−1]$  
第二步：将 $word1[i-1]$ 替换成 $word2[j-1]$   -->   $+1$

删除：  
删除 $word1[i-1]$ 这个字符  
第一步：将 $word1[:i-1]$ 变成 $word2[:j]$   
第二步：删除 $word1[i-1]$ 这个字符

插入：   
在 $word1$ 中插入 $word2[j-1]$ 这个字符  
第一步：将 $word1[:i]$ 变成 $word2[:j-1]$  
第二部：插入 $word2[j-1]$ 这个字符
