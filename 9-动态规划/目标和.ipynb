给你一个非负整数数组 nums 和一个整数 target 。

向数组中的每个整数前添加 '+' 或 '-' ，然后串联起所有整数，可以构造一个 表达式。
返回可以通过上述方法构造的、运算结果等于 target 的不同 表达式 的数目。

### 方法一：回溯
但是会超时

```python
class Solution:
    def findTargetSumWays(self, nums: List[int], target: int) -> int:
        self.count = 0
        self.backtrack(nums, target, 0, 0)
        return self.count
        

    def backtrack(self, nums, target, i, cur_sum):
        if i == len(nums):
            if cur_sum == target:
                self.count += 1
        else:
            self.backtrack(nums, target, i + 1, cur_sum + nums[i])
            self.backtrack(nums, target, i + 1, cur_sum - nums[i])       
```

count的前面需要加上self成为<mark>实例变量</mark>，
这样才可以在整个实例内共享

不加self的话是局部变量，只在当前方法调用内有效

### 方法二：动态优化 记忆化搜索


```python
class Solution:
    def findTargetSumWays(self, nums: List[int], target: int) -> int:
        size = len(nums)
        table = dict()  # 记忆化缓存：key=(索引, 当前和), value=方案数

        def dfs(i, cur_sum):
            # 1. 递归终止条件：处理完所有数字
            if i == size:
                return 1 if cur_sum == target else 0
            
            # 2. 记忆化查找：如果已经计算过这个状态，直接返回
            if (i, cur_sum) in table:
                return table[(i, cur_sum)]
            
            # 3. 递归计算：当前数字可以选择加或减
            cnt = dfs(i + 1, cur_sum - nums[i]) + dfs(i + 1, cur_sum + nums[i])
            
            # 4. 缓存结果，避免重复计算
            table[(i, cur_sum)] = cnt
            return cnt

        return dfs(0, 0)
```

`dfs(i, cur_sum)` 表示：从索引 i 开始，当前累积和为 cur_sum 时，能到达目标 target 的方案数

`table[(i, cur_sum)] = dfs(i + 1, cur_sum - nums[i]) + dfs(i + 1, cur_sum + nums[i])`: 当前状态 (i, cur_sum) 的方案数 = 给第 i 个数字加负号的方案数 + 给第 i 个数字加正号的方案数。因为下一步选择不是加就是减
