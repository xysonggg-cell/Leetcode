### 方法一：动态优化

```python
class Solution:
    def lenLongestFibSubseq(self, arr: List[int]) -> int:
        size = len(arr)
        dp = [[0 for _ in range(size)] for _ in range(size)]  # 二维DP表: 因为需要同时记录最后两个元素，才能推导出下一个元素
        ans = 0

        # 任何两个不同的元素都可以构成长度为 2 的斐波那契子序列
        for i in range(size):
            for j in range(i + 1, size):
                dp[i][j] = 2  # 初始化为 2

        idx_map = {}
        for idx, value in enumerate(arr):
            idx_map[value] = idx

        for i in range(size):
            for j in range(i + 1, size):
                # 查找是否存在下一个元素，使得 arr[i] + arr[j] = arr[k]
                if arr[i] + arr[j] in idx_map:
                    k = idx_map[arr[i] + arr[j]]  # 获取下一元素的索引
                    
                    # 状态转移：以 (j,k) 结尾的序列长度 = 以 (i,j) 结尾的序列长度 + 1
                    dp[j][k] = max(dp[j][k], dp[i][j] + 1)
                    ans = max(ans, dp[j][k])

        if ans >= 3:
            return ans
        return 0
```

状态定义：$dp[j][k]$ 表示以 $arr[j]$ 和$ arr[k]$ 结尾的最长斐波那契子序列长度

状态转移：如果存在$k$ 使得$arr[i]+arr[j]=arr[k]$，则可以在以 $arr[i]$、$arr[j]$ 结尾基础上接上 $arr[k]$，此时有 $dp[j][k]=dp[i][j]+1$。因此，状态转移方程为：$dp[j][k]=max⁡(dp[j][k],dp[i][j]+1)$

### 方法二：哈希表

```python
class Solution:
    def lenLongestFibSubseq(self, arr: List[int]) -> int:
        size = len(arr)
        table = dict()
        ans = 0
        for index, num in enumerate(arr):
            table[num] = index
        
        for i in range(size):
            for j in range(i + 1, size):
                temp_ans = 0
                temp_i = i
                temp_j = j
                while arr[temp_i] + arr[temp_j] in table:
                    k = table[arr[temp_i] + arr[temp_j]]
                    temp_ans += 1
                    temp_i = temp_j
                    temp_j = k
                
                ans = max(ans, temp_ans)

        if ans > 0:
            return ans + 2
        else:
            return ans
```
