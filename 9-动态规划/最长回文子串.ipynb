给你一个字符串 s，找到 s 中最长的回文子串

### 方法一：双指针

```python
class Solution:
    def longestPalindrome(self, s: str) -> str:
        longest = ""
        for i in range(len(s)):
            palindrome1 = self.isPalindrome(s, i, i)  # self调用类内部的函数，
                                                      # 如果没有 self，Python会去全局作用域找 isPalindrome 函数，但找不到就会报错
            palindrome2 = self.isPalindrome(s, i, i + 1)
            
            longest = longest if len(longest) > len(palindrome1) else palindrome1
            longest = longest if len(longest) > len(palindrome2) else palindrome2

        return longest

    def isPalindrome(self, s, left, right) -> str:
        while left >= 0 and right <= len(s) - 1:
            if s[left] == s[right]:
                left -= 1
                right += 1
            else:
                break
        return s[left + 1: right]
```

也是左右指针的思路，但这道题是让指针从中间到两端移动

注意在写函数和调用函数的时候不要忘了self

### 方法二：动态规划

```python
class Solution:
    def longestPalindrome(self, s: str) -> str:
        n = len(s)
        if n <= 1:
            return s

        dp = [[False for _ in range(n)] for _ in range(n)]
        max_start = 0
        max_len = 1

        for j in range(1, n):
            for i in range(j):
                if s[i] == s[j]:
                    if j - i <= 2:
                        dp[i][j] = True
                    else:
                        dp[i][j] = dp[i + 1][j - 1]
                if dp[i][j] and (j - i + 1) > max_len:
                    max_len = j - i + 1
                    max_start = i
        return s[max_start: max_start + max_len]
```

$dp[i][j]$ 表示为：字符串 $s$ 在区间 $[i,j]$ 范围内是否是一个回文串

状态转移方程：
当子串只有 1 位或 2 位的时候，如果 $s[i]==s[j]$，该子串为回文子串，即：$dp[i][j] = (s[i] == s[j])$。
如果子串大于 2 位，则如果 $s[i+1...j−1]$ 是回文串，且 $s[i]==s[j]$，则 $s[i...j]$ 也是回文串，即：$dp[i][j] = (s[i] == s[j])$ and $dp[i + 1][j - 1]$。
