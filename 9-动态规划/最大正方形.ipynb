在一个由 '0' 和 '1' 组成的二维矩阵内，找到只包含 '1' 的最大正方形，并返回其面积。

```python
class Solution:
    def maximalSquare(self, matrix: List[List[str]]) -> int:
        m = len(matrix)
        n = len(matrix[0])
        dp = [[0 for i in range(n)] for i in range(m)]
        max_side = 0

        for i in range(m):
            if matrix[i][0] == '1':
                dp[i][0] = 1
                max_side = max(max_side, dp[i][0])

        for j in range(n):
            if matrix[0][j] == '1':
                dp[0][j] = 1
                max_side = max(max_side, dp[0][j])
        
        for i in range(1, m):
            for j in range(1, n):
                if matrix[i][j] == '1':
                    dp[i][j] = min(dp[i - 1][j - 1], dp[i - 1][j], dp[i][j - 1]) + 1
                max_side = max(max_side, dp[i][j])
        
        return max_side ** 2
```

定义状态：  
$dp[i][j]$ 表示以 $(i,j)$ 为右下角、且全部为 1 的最大正方形的边长

状态转移方程：  
仅当 $matrix[i][j]==1$ 时，$(i,j)$ 位置才能作为正方形的右下角;  
如果 $matrix[i][j]==0$，则 $dp[i][j]=0$；  
如果 $matrix[i][j]==1$，则 $dp[i][j]=min(dp[i−1][j−1],dp[i−1][j],dp[i][j−1])+1$，即取其上方、左方、左上方的最小值加 1。
